### Desafio 01: Investigar por que a classe tratamento é tão desbalanceada?

````python
# Verificando a relação do tratamento com relação a dose e tempo
pd.crosstab([dados.tratamento,dados.tempo],
            dados.dose,
            margins=True)
````

````python
# Verificando a relação do tratamento com relação a dose e tempo
pd.crosstab(dados.tempo,
            [dados.tratamento, dados.dose],
             margins=True)
````

### Desafio 07: Deixar os gráficos bonitões. (Matplotlib.pyplot)

````python
# Bibliotecas para customizar os gráficos
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import rc, rcParams

# Define o estido do gráfico
sns.set_theme(style="darkgrid")
````

````python
# Customizando o gráfico de tratamento
fig, ax = plt.subplots(1, 1, figsize=(8,8))

# Plot do gráfico
sns.countplot(data=dados, x = 'tratamento', saturation=0.9, alpha=0.7, palette='YlGnBu')

# Formatação do título
plt.title('Balanceamento da classe tratamento', fontdict={'fontsize': 20}, weight='bold')

# Configuração geral da fonte que será utilizada nos eixos
font = {'fontsize':15, 'weight':'bold'}
plt.tick_params(axis='both', labelsize=15)

# Formatação do eixo x
plt.xlabel('Tratamento', fontdict=font)

# Formatação do eixo y
plt.ylabel('Quantidade de tratamento', fontdict=font)

fig.tight_layout()
plt.show()
````
